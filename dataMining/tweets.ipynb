```python
### pretrqitement des donnees
```

```python
import re
for index, row in twitter_data.iterrows():
    old = row['text']
    new0 = re.sub(r"http\S+", "", old)
    new1 = re.sub(r"#+","#", new0)
    new2 = re.sub(r"@", "", new1)
    new3 = re.sub(r"http\S+", "", new2)
    new4 = re.sub(r"[-()\"#/@;:<>{}+=~|.?,]", "", new3)
    twitter_data.loc[index,'text'] =new4
```
