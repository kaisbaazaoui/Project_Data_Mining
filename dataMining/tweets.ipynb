```python
### pretrqitement des donnees
```

```python
import re
for index, row in twitter_data.iterrows():
    old = row['text']
    new0 = re.sub(r"http\S+", "", old)
    new1 = re.sub(r"#+","#", new0)
    new2 = re.sub(r"@", "", new1)
    new3 = re.sub(r"[-()\"#/@;:<>{}+=~|.?,]", "", new2)
    twitter_data.loc[index,'text'] =new3
```

```python
### traitement des donnees avec NLP
```

```python
import nltk
#nltk.download('all')
```

```python
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords  
ps = PorterStemmer() 
stemed_dataset=[]
for i in range(0,twitter_data.shape[0]):
    stemmed_array=twitter_data['text'][i].split()
    stemmed=[ps.stem(word) for word in stemmed_array if not word in set(stopwords.words('english'))]
    stemmed=' '.join(stemmed)
    stemed_dataset.append(stemmed)
print(stemed_dataset[0:10])
```

```python

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X=cv.fit_transform(stemed_dataset)
print(X)
```

```python
print(twitter_data.iloc[0])
```

```python
### classification des tweets
from sklearn.cluster import KMeans
```

```python

num_clusters = 5
kmeans = KMeans(n_clusters=num_clusters,
                random_state=0, 
                n_init=10,
                init='k-means++',
                verbose=0, 
                max_iter=300).fit(X)

print ("clusters: ", num_clusters, "\tcluster inertia: ", kmeans.inertia_)

print ("cluster centers:\n", kmeans.cluster_centers_)
#open('cluster_centres.csv', 'w').writelines(','.join(map(str, kmeans.cluster_centers_.tolist())))


cluster_labels = kmeans.labels_
print(len(cluster_labels))
```
